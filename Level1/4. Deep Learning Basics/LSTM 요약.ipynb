## LSTM
- $X_t$ : input, 단어, word embedding 등..
- $h_t$ : output, hidden state
- **Previous cell state** : 네트워크 내부에서만 흐름. 타임스텝 $t$까지의 정보를 취합해서 summarize해주는 정보다.
- **Previous hidden state** : 다음번 $t+$1번째 LSTM의 Previous hidden state가 된다. (Next hidden state)

### 3가지 Gate
- Forget gate
- Input gate
- Output gate

### ```Forget Gate```
- 어떤 정보에서 어떤 것을 살리고 버릴 지를 결정한다.
- 이전 값인 Previous hidden state($h_{t-1}$)와 현재 입력($X_t$)을 이용한다.

### ```Input Gate```
- 어떤 정보를 현재 셀 state에 올릴 지(추가할 지)를 결정한다. (이전 정보를 무작정 현재 셀에 올리는게 아님)
- 이전 값인 Previous hidden state($h_{t-1}$)와 현재 입력($X_t$)을 이용한다.
- **```Update Cell```**를 통해 업데이트를 한다.
- $\tilde{C_t}$ : 이전 cell state와 현재 입력이 들어와서, tanh을 통해 계산된다. 즉, 현재 정보와 이전 출력 값을 가지고 만든 cell state의 예비군이다.
- 버릴거 버리고, 현재 정보를 기준으로 쓸거는 쓰고, 이런식으로 합쳐서 새로운 cell state $C_t$를 업데이트하게 된다.

### ```Output Gate```
- 어떤 값을 밖으로 내보낼 지를 결정한다.
- 현재 output과 next hidden state로 나가게 된다.
