## plt 그래프

```python
fig = plt.figure()

x1 = [1,2,3]
x2 = [3,2,1]

ax = fig.add_subplot(121)
plt.plot(x1)
ax = fig.add_subplot(122)
plt.plot(x2)
plt.show()
```

### 그래프 색상

```python
fig = plt.figure()
ax = fig.add_subplot(111)

# 문자, 컬러네임, hex code로 색상 지정이 가능함
ax.plot([1,1,1], color='r')
ax.plot([2,2,2], color='forestgreen')
ax.plot([3,3,3], color='#000000')
plt.show()
```

### 텍스트 사용하기

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1,1,1], label='1')
ax.plot([2,2,2], label='2')
ax.plot([3,3,3], label='3')

# 그래프의 제목을 추가
plt.title('Basic Plot')

# 범례를 추가해서 label을 보여줌
plt.legend()
plt.show()
```

```python
# subplot이 있을때, 전체에 대해 제목 달기
# super title -> suptitle
fig = plt.figure()

ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

# 각 subplot마다 제목을 달 수 있다.
ax1.set_title('ax1')
ax2.set_title('ax2')

# fig라는 전체 공간에 제목을 달기 위해 suptitle 사용
fig.suptitle('fig')

plt.show()
```

```python
# set 메소드로 만들어진 정보들은
# get 메소드로 정보를 받을 수 있다.
ax.get_title()
```

### 축 (ticks)

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1,1,1], label='1')
ax.plot([2,2,2], label='2')
ax.plot([3,3,3], label='3')

ax.set_title('Basic Plot')
# x축을 지정할 수 있음
ax.set_xticks([0,1,2])
# 0 1 2로 지정한 x축을 label로(텍스트로) 바꿀 수 있음
ax.set_xticklabels(['zero','one','two'])

# 범례를 추가해서 label을 보여줌
plt.legend()
plt.show()
```

### 일반적인 텍스트를 추가하는 방법
- text
- annotate

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1,1,1], label='1')
ax.plot([2,2,2], label='2')
ax.plot([3,3,3], label='3')

ax.set_title('Basic Plot')
ax.set_xticks([0,1,2])

# 위치를 지정하고 텍스트를 적을 수 있음
ax.text(x=1, y=2, s='this is text')

# 주석으로 원하는 텍스트를 추가할 수 있음, xy위치는 튜플형태로 입력
# annotate는 화살표 추가가 가능
ax.annotate(text='this is text2', xy=([1,3]),
            xytext=(1.2, 2.7),
            arrowprops=dict(facecolor='black'))

# 범례를 추가해서 label을 보여줌
plt.legend()
plt.show()
```
