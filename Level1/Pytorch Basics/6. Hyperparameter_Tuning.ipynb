# 하이퍼 파라미터
- 모델 스스로 학습하지 않는 값
- 사람이 지정해야 되는 값들을 의미함
- ex) learning rate, 모델의 size, optimizer 등..

### 성능을 올리는 방법 
1. 모델을 바꿔서 성능을 올린다.
2. 데이터를 바꿔서 성능을 올린다. (새로운 데이터를 추가하거나 기존 데이터의 오류를 고침)
3. 하이퍼 파라미터를 튜닝한다.


- 사실 2번(데이터)이 성능을 올리는 가장 좋은 방법이다.
- 데이터는 다다익선이다.
- 3번으로 얻는 이익은 생각보다 크진 않음. 마지막으로 성능을 0.01이라도 올리기 위해 사용되는 경우가 많음

### 튜닝 방법
1. Grid Search
    - 값들의 일정한 범위를 정해서 학습을 수행함
    - 주로 $log$ 단위로 올림 (0.1, 0.01, 0.001...)


2. Random Search
    - 랜덤한 값으로 학습을 수행함


3. 요즘에는 베이지안 기반의 기법들이 유행임. (BOHD)

## Ray
- multi-node multi processing을 지원하는 모듈이다.(여러대 컴퓨터를 연결함)
- 기계학습의 병렬 처리를 위해 개발된 모듈임.
- 하이퍼파라미터 서칭을 위한 다양한 모듈이 있다.
