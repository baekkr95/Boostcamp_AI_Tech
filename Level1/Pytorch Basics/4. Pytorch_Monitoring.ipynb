## Tensorboard
- 텐서플로우의 프로젝트로 만들어진 시각화 도구이다.
- 학습 그래프, metric, 학습 결과 등 여러 시각화가 가능함
- 파이토치도 연결이 가능함

### 저장하는 값
- scaler : metric(정확도, 손실, precision, recall 등..) 상수 값의 epoch을 표시
- graph : 모델의 computational graph 표시
- histogram : weight 등 값의 분포를 표현해줌
- image, text : 예측 값과 실제 값 비교에 사용
- mesh : 3d 형태의 데이터를 표현해주는 도구

1. SummaryWriter을 만들고
2. 기록할 위치를 지정하고
3. wrtie.add_(타입) 으로 값을 추가하고, flush(=close) 저장 완료

### 텐서보드 띄우기

1. ```%tensorboard --logdir {디렉토리}``` : 텐서보드 다운받고, 디렉토리 지정하기
2. ```%load_ext tensorboard``` : 주피터나 코랩에서 사용하는 방법

## wandb
- MLOps 도구
- 머신러닝 실험을 원활히 지원하기 위한 도구
- 협업, 실험 결과 기록, code versioning 등 여러 기능 제공

- wandb 설치
- config에 하이퍼 파라미터에 대한 정보를 적어준다.
- 마지막에 log를 통해 기록 저장 (텐서보드의 add와 동일)
