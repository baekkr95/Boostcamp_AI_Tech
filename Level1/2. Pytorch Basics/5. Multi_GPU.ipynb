## 멀티 GPU
- 대용량의 데이터를 학습하고, 성능이 좋은 모델을 만들기 위해 필요해짐
- 주로 Single Node (1대의 컴퓨터) Multi GPU 를 의미함.

### Model Parallel
- 다중 GPU에 학습을 분산하는 2가지 방법
    1. 모델을 나누는 방법
    2. 데이터를 나누는 방법


- 모델을 나누는 것은 예전부터 자주 사용했음 (AlexNet)
- 모델의 병목현상, 파이프라인의 어려움 등 때문에 여전히 모델 병렬화는 고난이도 과제다.

- AlexNet 중간 레이어를 보면 모델 병렬화가 나타난다는 것을 볼 수 있다.
- 이 당시 GPU 성능이 좋지 않아서 2개로 나눠서 처리했었다.

- 모델 병렬화를 잘못하면, 위 사진처럼 비는 공간이 생긴다.
- GPU 0과 GPU 1이 동시에 처리되는 파이프라인을 만들어야 함

### Data Parallel
- 데이터를 나눠 GPU에 할당하고, 결과의 평균을 취하는 방법이다. (DataParallel)
    - Gpu 사용 불균형 문제가 생긴다.
    - 하나의 GPU가 병목이 생긴다. (중앙에서 coordinate하는 GPU가 있음)
    - GIL (global interpreter lock)
    

- 미니배치와 유사한데, 한번에 여러 Gpu에서 수행을 한다.
- DistributedDataParallel를 사용하면 개별적으로 연산의 평균을 내기 때문에, 위에서 언급한 문제를 좀 덜 걱정해도 된다.
