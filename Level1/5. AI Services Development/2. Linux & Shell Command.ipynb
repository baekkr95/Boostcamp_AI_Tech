### 리눅스 배포판 (계보)


## Shell Command

### 쉘이란?
- 사용자가 문자를 입력해서 컴퓨터에 명령할 수 있도록 하는 프로그램
- 같이 나오는 개념이, 터미널과 콘솔이 있다
- 터미널/콘솔 : 쉘을 실행하기 위해 문자 입력을 받아 컴퓨터에 전달해줌. 프로그램의 출력을 화면에 작성해준다.

### 쉘 종류
- sh : 최초의 쉘
- bash: 리눅스 표준 쉘
- zsh : 맥 카탈리나 OS 기본 쉘

### 쉘을 사용하는 경우
- 서버에 접속해서 사용할 경우
- 리눅스 내장 기능을 사용할 경우 (예를 들면 crontab)
- 데이터 전처리를 위해 쉘 커맨드를 사용
- 도커를 사용할 경우
- 여러 서버를 관리할 경우
- Test Code 실행할 경우
- 배포 파이프라인 실행할 경우 (깃허브 Action 등에서 실행)

## 생활코딩 리눅스 유튜브 참고

### ls

- ls를 통해서 현재 위치에 있는 파일 및 디렉토리들을 확인
- ```-l``` 옵션을 통해 더 자세한 정보들을 알 수 있다.
- 맨 앞의 내용은 권한을 의미하는데, 앞에 d가 붙은 것은 디렉토리를 의미하고, d가 없으면 파일을 의미한다.

### rm
- 원하는 파일을 삭제할 수 있다.
- 하지만, 디렉토리를 삭제할 수 없다,
- ```-r``` 옵션을 붙여야 디렉토리를 삭제할 수 있다.

### 파일 편집기
1. nano (쉬움)
2. vim (부캠 강의에 나옴)

### wget
- url를 통해 파일을 다운도르 받는다
- ```-O``` 옵션을 통해 다운 받은 파일의 이름을 설정할 수 있음
- ```wget -O [원하는 이름] [다운로드 URL]```

## IO Redirection
### Redirection - output
- ```>```또는 ```1>``` 기호를 사용함 (standard output)
- (예시) ```ls -l > result.txt``` : ls -l라는 출력이 result.txt라는 파일에 저장됨.


- ```2>``` 기호를 사용함 (standard error)
- (예시) ```rm rename2.txt 2> result.txt``` : rm rename2.txt 명렁어에 error가 발생하면 error 출력이 result.txt라는 파일에 저장됨.

### Redirection - input (실용적이지 않음)
- ```<``` 기호를 사용함

### Redirection - append
- ```>>``` 기호를 사용함
- 내용을 추가할 수 있다. (같은 내용을 넣어도 덮어쓰기 안 되고 추가된다.)

## Shell Script
- 쉘에서 실행되는 명령어들을 어딘가에 적어두고
- 그것을 그대로 컴퓨터가 실행하게 한다.
- 여러가지 일들을 자동화시킬 수 있다.

- ```#!/bin/bash``` : bash라는 프로그램을 통해 해석된다 라는 약속임. 이 밑에 있는 코드들이 bash라는 문법에 따라 해석이 되고 실행되게 할 것임.
- 만약 bak 이라는 디렉토리가 없으면, 디렉토리를 만들고
- .log 확장자가 붙은 모든 파일을 bak 디렉토리로 복사하라는 것임

## 프로세스 확인
- ```ps```
- ```top``` : sudo top을 실행시키면 프로세스 리스트를 출력
- ```htop``` : sudo apt-get install htop으로 설치하고, sudo htop으로 출력

## 파일 위치 찾기
- ```locate```: 디렉토리가 아닌 데이터베이스에서 파일을 찾기 때문에 빠르다.


- ```find``` : 디렉토리를 뒤져서 파일을 찾음. 옵션, 사용방법이 다양함
    - **<예시>**
    - `find / -name *.log` : 컴퓨터의 모든 디렉토리를 뒤져서 log라는 파일을 찾아라.
    - 만약 permission denied 된 파일도 보려면, 명령어 맨 앞에 `sudo`를 넣는다.
    - `-exec` 옵션을 붙여서 원하는 파일을 찾고, 해당 파일에 대해 다양한 명령들을 추가로 사용할 수 있다.
    
    
- `whereis` : 원하는 파일의 위치를 출력함

## Permission (권한)
- 파일과 디렉토리에 대해서 Read, Write, Excute 에 대한 권한을 지정하는 것


### 나(k8805)는 egoing이라는 다른 사람의 파일(perm.txt)을 수정할 수 없다. (permission deined)

### Access Mode는 3가지 다시 구분된다.
- 첫번째 부분은 owner의 권한
- 두번째 부분은 group의 권한
- 세번째 부분은 other(불특정 사용자들)의 권한


- `r` : read
- `w` : write
- `x` : execute
- perm.txt의 파일에 대해서 나는(k8805)는 r(read)만 가능하기 때문에 수정이 안 됨

### Access Mode 바꾸기 (CHMOD)
- `chomd` 명령어를 사용
- 만약 egoing이 others가 perm.txt를 읽지도 못하게 하려면, `chmod o-r perm.txt`를 하면 된다.
- 만약 egoing이 others가 perm.txt를 읽을 수 있게 하려면, `chmod o+r perm.txt`를 한다.
