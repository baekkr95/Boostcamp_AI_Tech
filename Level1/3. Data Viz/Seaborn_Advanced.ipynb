````python
import numpy as np 
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

student = pd.read_csv('../../../data/exams.csv')
iris = pd.read_csv('../../../data/iris.csv')
````

## Joint Plot
- 결합확률 분포를 볼 수 있는 그래프
- 결합확률 분포 뿐만 아니라 단일 변수의 분포도 볼 수 있음
- kind 파라미터를 통해 다양한 그래프를 활용할 수 있음

````python
sns.jointplot(x='math score', y='reading score',data=student,
             height=6)
````

````python
# hue를 통해 구분할 수 있음
sns.jointplot(x='math score', y='reading score',data=student,
              hue='gender'
             )
````

````python
# kind를 통해 다양한 그래프를 이용해서 joint plot를 그릴 수 있다.
sns.jointplot(x='math score', y='reading score',data=student,
              hue='gender', 
              kind='hist', # { “scatter” | “kde” | “hist” | “hex” | “reg” | “resid” }, 
              fill=True
             )
````

## Pair Plot
- 데이터셋의 pair-wise 관계를 시각화 할 수 있다.

- `kind`는 전체 서브플롯, `diag_kind`는 대각 서브플롯을 조정합니다.

- `kind` : {‘scatter’, ‘kde’, ‘hist’, ‘reg’}
- `diag_kind` : {‘auto’, ‘hist’, ‘kde’, None}

````python
# 마찬가지로 hue로 데이터를 구분해서 시각화할 수 있다.
sns.pairplot(data=iris, hue='Species')
````

````python
# joint plot처럼 다양한 그래프를 활용할 수 있다.
sns.pairplot(data=iris, hue='Species', diag_kind='hist')
# sns.pairplot(data=iris, hue='Species', kind='hist')
````

````python
# 대각선을 기준으로 대칭이니까 한쪽만 볼 수 있다.
sns.pairplot(data=iris, hue='Species', corner=True)
````

## Facet Grid
- pair plot이 변수(feature)들 간에 관계를 보여준다면, facet grid는 feature's category들 간에 관계를 볼 수 있다.

> #### 4가지 함수를 사용할 수 있음
- ```catplot``` : Categorical 
- ```displot``` : Distribution
- ```relplot``` : Relational
- ```lmplot``` : Regression
    

### Facet Grid - Cat Plot

````python
# 기본은 strip plot이다.
sns.catplot(x="race/ethnicity", y="math score", hue="gender", data=student)
````

````python
# 마찬가지로 kind 파라미터를 통해 다른 그래프를 사용할 수 있다
sns.catplot(x="race/ethnicity", y="math score", hue="gender", data=student,
            kind='box'
           )
````

````python
# 행과 열의 category를 기반으로 해당 그래프의 개수가 정해진다.
sns.catplot(x="race/ethnicity", y="math score", hue="gender", data=student,
            kind='box', 
            col='lunch', row='test preparation course'
           )

print(student.lunch.unique())
print(student['test preparation course'].unique())
````

- `lunch`와 ```test preparation course```가 각각 2개의 카테고리를 갖는다,
- 따라서 4개의 그래프가 생김

### Facet Grid - Dist Plot

> kind 파라미터를 사용해서 3가지 그래프를 사용할 수 있음
- `histplot()` (with `kind="hist"`; the default)
- `kdeplot()` (with `kind="kde"`)
- `ecdfplot()` (with `kind="ecdf"`; univariate-only)

````python
sns.displot(x="math score", hue="gender", data=student)
````

````python
# 마찬가지로 kind 파라미터를 사용할 수 있음
sns.displot(x="math score", hue="gender", data=student,
           col='race/ethnicity', # kind='kde', fill=True,
            col_order=sorted(student['race/ethnicity'].unique())
           )
````

### Facet Grid - relplot

````python
sns.relplot(x="math score", y='reading score', hue="gender", data=student,
           col='lunch')
````
