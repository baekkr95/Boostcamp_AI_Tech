## Segmentation FCN

### FCN 주요 특징
1. VGG를 백본으로 사용해서 feature extracting을 함
2. FC layer를 Convolution으로 대체함
3. Transposed Convolution으 이용해서 Pixel Wise prediction을 수행함.

- 각 픽셀마다 어떤 클래스를 가지는지에 대한 정보가 중요해서 위치정보를 유지하는게 중요하다.
- FC layer는 Flatten을 하기 때문에 위치정보를 해치니까 Fully Convolution Network을 사용하게 됐다.
- 또한 임의의 입력값에 대해서도 상관없이 학습과 inference가 가능하다.
- kernel의 파라미터에 의해 영향을 받고, 이미지 크기에 대해서는 상관이 없다.

- Transposed Convolution을 통해 Upsampling을 하게 된다.
- 이때 사용되는 Transposed Convolution Kernel 또한 일반적인 convolution kernel 처럼 학습이 되는 kernel이다.
- 위의 그림처럼 upsampling이 진행된다.
- 겹치는 영역이 생기면 Sum을 통해 합쳐주면 된다.


- 만약 output이 4x4가 아니라 5x5로 만들고 싶으면 stride를 2로 바꾸면 된다.
- 만약 padding을 주면 외곽 영역을 지운다.

- 기존의 FCN의 Transposed Convolution은 한번에 너무 크기를 키우기 때문에 성능이 정확하지 않음 (stride가 너무 크다)
- 그래서 skip connection을 사용해서 stride를 줄이게 된다.

- 실제로 32s에서 8s로 갈수록 성능이 좋아지고 디테일도 살릴 수가 있다.
