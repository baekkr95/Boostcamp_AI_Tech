### FCN 한계점
1. 객체가 크거나 작으면 예측을 잘 못한다. 이미지가 크면 지역적인 정보만으로 예측을 하거나, 이미지가 작으면 객체가 무시되는 문제가 있다.
2. 객체의 디테일한 모습이 사라지는 문제가 발생한다. Deconvolution의 절차가 간단해서 나타나는 현상이다.


이를 개선하고자 Decoder를 개선하거나, Receptive field를 개선하는 등 다양한 개선법들이 생기게 된다.

---
## Decoder를 개선한 모델
- DeconvNet
- SegNet

### DeconvNet

- 인코더(Convolution Network)와 디코더(Deconvolution Network)를 대칭해서 모델 구조를 만들었다.
- 디코더 부분 : Unpooling과 Transposed Convolution이 반복적으로 이루어짐
- Unpooling을 통해 Pooling시에 지원진 경계에 대한 정보를 기록했다가 복원을 해주는 역할을 한다. 그래서 객체의 디테일한 모습을 살릴 수 있다.
- 하지만 Unpooling을 하게되면 거의 0으로 값을 채우기 때문에 Transposed Convolution을 통해 dense하게 값을 채워야 한다.

### SegNet

- 성능보다는 속도 측면에서 많은 개선을 이룸.
- DeconvNet 중앙의 7x7과 1x1 Conv, Deconv를 없앰으로써 파라미터를 줄였다.
- Transposed Convolution 대신에 Unpooling만 사용해서 학습을 하지 않아도 되게 했다.
- 마지막 Score를 3x3 conv로 계산을 한다.

## Receptive Field를 확장시킨 모델
1. DeepLab v1
2. DilatedNet


- maxpooling을 사용하면 receptive field가 넓어지는 효과가 있다.
- 하지만 너무 자주하면 해상도 문제가 생긴다. 그렇다고 안 쓰면 파라미터가 너무 많아진다.
- 그래서 이미지 크기도 적게 줄이면서 receptive fields도 넓히기 위해서 dilated convolution이 나오게 됐다.


### DeepLab v1


- 2x2 maxpooling을 3x3 maxpooling으로 바꿈.
- dilated convolution을 적용했다.
- Upsampling에서 bilinear interpolation을 사용함
- Dense CRF 후처리 작업을 통해 좀 더 디테일을 살림.

### DilatedNet


- 3x3 대신에 2x2 MaxPooling을 적용함
- 4번과 5번 Layer에 Pooling 자체가 사라짐
- dilated convolution을 적용할 때, 패딩을 사용해서 이미지의 사이즈를 변경하지 않는 것이 중요하다. (3x3에서 dilated rate와 padding의 값이 동일하면 이미지 사이즈가 안 변함)
- Upsampling에서 Deconvolution을 사용했다.
- Context Module을 제안했음.
