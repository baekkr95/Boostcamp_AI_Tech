Semantic Segmentation보다 좀 더 진보된 Segmentation 방법인 Instance와 Panoptic Semgentation이 있다.  
단순히 픽셀 마다의 클래스를 분류하는 semantic segmentation은 동일한 클래스에 속하는 개별 물체를 구분하지 못합니다. 이와 달리 instance segmentation은 영상 내에 동일한 물체가 여러 개 존재하는 경우에 각각의 물체를 구분하며 동시에 픽셀 단위의 mask도 예측하는 방법입니다.  
그리고 semantic segmentation과 instance segmentation을 결합하여 더욱 복잡한 task인 panoptic segmentation이 있습니다.

## Instance Segmentation
- Semantic과 다르게, 같은 물체(class)라도 인스턴스가 다르면 다르게 segmentation을 한다.
- 그래서 Objective Detection 모델을 기반으로 하는 경우가 많음.

### Mask R-CNN
- Faster R-CNN을 개선함.
- 2 stage model 이다.
- 기존에는 RPN을 통해 나온 bbox proposals에 RoI Pooling을 사용했기 때문에 정수좌표가 뽑혔다.
- Mask R-CNN에서는 RoI Align을 사용해서 소수점까지 뽑기 때문에 정교한 feature를 뽑을 수 있게 된다. (interpolation 사용)

- Mask Branch라는 새로운 branch를 추가했음.
- class와 box를 예측하는 것은 faster R-CNN과 비슷하고, 
- mask branch는 7x7에서 14x14로 업샘플링하고 채널 수를 줄인다.
- 그 다음에 각 class 별로 (80개) Binary mask를 prediction 한다.
    - 하나의 bbox에 대해서 일괄적으로 모든 class에 대한 mask를 생성함.

### YOLACT
- 1 stage model, real time으로 instance segmentation이 가능함.


- Feature Pyramid 구조를 사용함. (FPN, feautre pyramid network)
    - 고행상도의 feature map를 가지고 사용할 수 있음
    - FPN 관련 글 : https://eehoeskrap.tistory.com/300


- 마스크의 prototype를 추출해서 사용한다.
    - mask r-cnn에서는 80개의 class를 고려하고 있다. 그래서 bbox마다 80개씩 생성하고 classification 결과에 따라 하나를 참조한다.
    - 하지만 여기선 마스크는 아니고, 마스크를 합성해낼 수 있는 기본적인 soft segmentation components라고 생각하면 된다.
    - 추후에 마스크로 합성될 수 있는 재료를 제공하는 것.
    
    
- Prediction Head는 각 detection에 대해서 프로토타입들 잘 합성할 수 있도록 계수(coeffcients)를 출력한다.
    - 이 계수들과 프로토타입을 선형 결합한다.
    - 최종적으로 각 dectection에 적합한 mask response map를 생성한다. (우상단 그림)
    

- object 개수에 상관없이 (굳이 80개를 고려하지 않아도 됨) 적당하게 설정하고, 그것의 선형결합으로 다양한 마스크를 생성하는 것이 key point
- 메모리적으로 효율성이 증가함

## Panoptic Segmentation
- Semantic은 배경에는 관심이 없고, 물체에만 segmentation을 진행한다.
- Panoptic은 배경 정보 뿐만 아니라, 관심있는 물체에도 instances segmentation을 진행함.
- 배경(Stuff), 물체(Things)

### UPSNet
- 여기서도 FPN 구조를 사용함. 고해상도의 feature map를 뽑을 수 있다.


- 그 다음 head branch를 여러개로 나눔
1. Semantic Head
    - fully convolution 구조로 되어서, semantic map를 prediction한다.
2. Instance Head
    - 물체의 detection과 box의 regression, 그리고 마스크를 추출하는 mask logits으로 나뉜다.
    

- 맨 뒷단에 위의 개별적인 task들을 융합해주는 Panoptic Head가 들어가서 최종적으로 단 하나의 segmentation map으로 합쳐준다.
