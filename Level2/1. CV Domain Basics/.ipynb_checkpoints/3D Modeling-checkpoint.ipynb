#### 3D가 활용될 수 있는 분야
1. AR, VR
2. 3D Printing
3. Medical application (예를 들면 단백질 sequence를 예측하는 것)


이미지란 3D가 투영된 2D 공간을 보는 것이다. 근데 만약에 3D를 projection한 2D사진 2개가 있으면 3D로 재구성할 수가 있다.  
Triangulation : 2D 이미지로부터 3D를 구하는 방법 중 하나

### 3D를 어떻게 CV로 표현할까?
- 2D 이미지는 2D array, RGB 값으로 표현할 수 있다.
- 3D는 unique하게 표현할 수 없다.
    - Multi view images (여러 각도로 사진을 촬영해서 복원하는 방법)
    - Volumetric (2D와 가장 비슷하게 표현)
    - Part assembly (도형들의 parametric한 집합으로 표현)
    - Point cloud (3D 상의 포인트들의 집합, 하나의 점이 x,y,z 좌표를 갖는다)
    - Mesh, Graph CNN (3D 포인트들이 공간 상에 있으면, 삼각형 형태로 이어진다)
    - Implicit shape (고차원의 함수로 표현)

## 3D Dataset
### ShapeNet
- 대표적인 3D 이미지 데이터셋
- 51300개의 데이터와 55개의 카테고리로 나눠진 데이터셋이다.
- 2D의 ImageNet과 같다고 생각하면 된다.


### PartNet
- ShapeNet의 개선 버전
- Fine-grained dataset : 하나의 물체에 대해서 detail들이 annotation 되어있음
- 57만개 정도의 part instances을 구분했고 (segmentation에 유용함)
- 2만 6천개 정도의 3D 데이터가 있다.



### SceneNet
- 500만개의 RGB + Depth pair의 영상 데이터셋
- indoor images를 사용함
- indoor image들을 3D 스캔하고 랜덤하게 generation해서 시뮬레이션한 이미지임.




### ScanNet
- 마찬가지로 RGB, Depth pair로 이루어진 데이터셋
- 250만개의 데이터가 존재함
- SceneNet과 달리 시뮬레이션이 아닌 실제 indoor를 스캔한 이미지임




### Outdoor 3D datasets
- 자율 주행차를 위해 준비한 데이터셋으로 보면 된다.
- KITTI, Waymo open dataset 등이 있다.

## 3D Recognition
- 2D CNN을 사용해서 단순히 label를 출력하는 것과 비슷한 과정을 갖는다.
- 마찬가지로 3D CNN를 사용해서 label를 예측 출력한다.

## 3D Object Detection
- 이미지 또는 3D 공간에서 Localization을 한다.
- 무인차 어플리케이션에서 유용하게 사용한다.

## 3D semantic segmentation
- 물체에 대해서 각 파트별로 segmentation을 한다.

## Conditional 3D generation
- Mesh R-CNN 모델을 사용한다.
- 2D를 입력으로, 3D을 출력으로 한다.
- detection도 동시에 가능하다.
- Mask R-CNN에 3D branch (3D Mesh)가 추가된 형태를 갖는다.

## Photo refocusing
- 하나의 이미지를 depth map를 이용해서 refocusing 또는 defocusing 하는 것

### Defocusing 기능 구현 방법
1. focus를 맞추고 싶은 Depth threshold를 정한다.
    - min depth ~ max depth 영역에 focus를 할지 말지 결정
    - 또는 max depth 이후로는 focus를 전부 날린다




2. depth map를 thresholding해서 mask를 2개 생성함.
    - max를 넘어가면 defocusing area
    - threshold 안에 포함되면 focusing area

3. 입력 이미지의 blurred 버전을 만든다.
    - blur kernel를 사용한다.
    - kernel 사이즈는 임의로 정해도 된다.
    

4. Original 이미지에 focusing area mask를 적용하고, 3번의 blurred 이미지에 defocusing area mask를 적용해서 추출한다.
    - 그럼 focus image는 선명하고 샤프한 이미지가 나오고
    - defocus 이미지는 흐릿하게 나온다.



5. 그 후 2개를 blending해서 refocused image가 만들어진다.
