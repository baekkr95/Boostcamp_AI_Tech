## Landmark Localization
- 주로 얼굴이나, 포즈를 tracking 하는데 사용한다.
- 중요한 key points를 landmark로 정의하고, 그것들을 추정하는 것

### Heatmap classification
- x,y 위치를 예측하는 `Coordinate Regression` 방법이 있지만, 이 방식은 정확하지도 않고 일반화에도 문제가 있어서 그 대안으로 Heatmap 방식이 나왔음.
- 성능은 훨씬 좋아졌지만, 모든 픽셀에 대해 판별을 해야돼서 계산량이 많아졌음.


- 각 채널들이(N개) 각각의 key points가 된다.
- 각 key point마다 하나의 class로 생각해서, key point가 발생할 확률 map을 각 픽셀별로 classification 하는 방법으로 해결함.
- 각 위치에 해당하는 landmark가 어떤 확률 map처럼 표현되는 것을 볼 수 있다.

### Hourglass Network
- U-Net의 구조를 쌓아놓은 모양이다.


- 영상 전체를 작게 만들어서, receptive field를 키운다. (전반적으로 큰 그림을 보자)
- 그것을 기반으로 landmark를 찾는 것임.
- 중간중간 skip connection이 있어서, low level feature를 참고해서 정확한 위치를 특정할 수 있게 만들엇다.


- 위의 과정을 한번에 하지 않고, 계속 반복하면서 점점 더 큰 그림과 디테일을 구체화하면서 refine하는 방식이다.

#### Hourglass 구조
- U-Net하고 비슷하다.
- 다른 점은, 뒷 부분에 $+$ 기호로 되어있다. U-Net은 concat을 하지만, hourglass는 더하기 때문에 차원이 증가하지 않는다.
- 대신, skip connection 중간에 또 다른 conv layer를 통과해서 전달된다. 이 부분 때문에 FPN과 닮았다고도 한다.

## DensePose
- UV map을 사용
    - 표준 3d 모델의 각 부위를 2d(U와 V축)으로 펼쳐서 이미지 형태로 만든 좌표 표기법
    

- Mask r-cnn과 비슷하게 생겼다. 쉽게 말하면 Faster R-CNN에 3D surface regression branch를 추가한 모델이다.
- Patch는 각 body part의 segmentation map 이다.
- 입력과 출력간의 관계를 잘 설명해서 2D 구조의 CNN으로 3D를 잘 예측하는 방법이다.
