- 주어진 데이터셋의 분포를 실제 데이터 분포와 최대한 유사하게 만들거나,
- 이미 학습된 정보를 이용해 새 데이터셋에 대해 보다 잘 학습하거나,
- label이 없는 데이터셋까지 이용해 학습하는 등 주어진 데이터셋을 최대한 효율적으로 이용해 딥러닝 모델을 학습하는 방법

## 1. Data Augmentation
- 우리의 train data들은 real data를 다 표현할 수 없다.
- 결국 sampling된 데이터이고 심지어 bias 되어있다
- real data와 sample data간에 갭을 채우고자 하는 것이 augmentation이다.

### Image data augmentation
- 밝기 조절
- 회전, 반전
- 자르기(crop)
- Affine transformation
- CutMix
- 여러 기법들을 opencv나 numpy에서 쉽게 사용이 가능함.
- 그리고 여러 기법들을 다양하게 조합해서 다양한 이미지를 생성할 수 있고, 이를 쉽게 도와주는 것이 RandAugment 알고리즘
- augment 조합과 강도를 파라미터로 사용한다.

## 2. Pre-trained
- 데이터를 모으고 label을 달아주는 것 자체가 너무 힘들다.
- 이를 해결하기 위해, transfer learning을 사용함.
- `transfer learning` : 한 데이터셋에서 배운 지식을, 다른 데이터셋에서 활용하는 것

### Transfer learning 방식
1. Feature Extractor : 학습된 모델의 fc layer를 새로운 task에 맞게 바꿈
2. Fine-tuning : 새로운 fc layer뿐만 아니라 conv layer도 같이 학습을 한다. (기존 conv layer는 learning rate를 낮게 설정함)

### Knowledge distillation
- transfer learning을 더 진보된 방법으로 사용함
- Teacher-student learning 이라고도 불림
  

- 이미 학습된 Teacher Network의 지식을 더 작은 모델인 Student Network에 주입해서 학습함.
- 모델 압축에 유용함.


- 최근에는, Teacher에서 생성된 출력을 unlabeled된 데이터의 가짜레이블로 자동 생성하는 메카니즘으로 사용함. `(pseudo-labeling)`
- 가짜 레이블로 자동 생성되면 더 큰 Student Network를 사용할때도 regularization 역할로 사용이 가능함.


#### 레이블 없는 경우

- 위 사진에서, 역전파는 Student에게만 간다. 즉, student만 업데이트가 된다.
- KL divergence는 2개 출력의 분포를 비슷하게 만들어 줌.
- 결국 student가 Teacher를 따라하게 함
- 그리고 label를 사용하지 않아서, 위의 경우에는 unsupervised learning이라 볼 수 있다.

#### 만약 레이블이 존재하는 데이터를 가지고 있을 경우

- Student Loss : True label(ground truth)을 사용한 loss (일반적으로 cross-entropy 사용)
- Distillation Loss : Teacher를 따라하게 만드는 Loss (KL Divergence)
- 최종적으로 Distillation Loss와 Student Loss의 Weighted Sum을 사용함.
- 그리고 역전파는 Student Model로만 흐르게 함.


- Distillation을 할 때는 Soft Prediction을 한다.
- Soft Prediction : 실수값을 가짐 (One-hot 벡터를 hard label이라고 한다.)


- Temperature(T) : 출력을 스무스하게 바꿔줌. (더 많은 정보를 사용하게 된다)
- 소프트맥스만을 취하면 입력값을 극단적으로 벌리게 된다.

## 3. Unlabeled dataset for Training
### semi-supervised learning with pseudo labeling

1. labeled data로 모델을 pre-train을 한다.
2. unlabeled data의 pseudo label을 해당 모델을 통해 자동 생성한다.
3. 그리고 2개의 dataset을 합쳐서 새로운 모델(또는 이전의 모델)을 재학습한다.

## 4. 최신 연구

### Self-training
- Augmentation + Teacher-student network + semi-supervised learning 을 전부 합침
- iteration이 진행됨에 따라, student model이 점점 커진다.
