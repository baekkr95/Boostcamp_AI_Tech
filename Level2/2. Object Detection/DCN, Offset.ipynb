일정한 패턴을 지닌 conv network는 geometirc transformation에 한계가 있다.
같은 물체여도 형태, 각도, 위치가 달라지니까 결과값이 달라짐.

그래서 정사각형의 일정한 패턴이 아닌, conv할 픽셀의 위치를 찾아주는 Offset이라는 개념과 offset를 학습할 것을 제안함.


Deformable Convolution을 제안함
어떤 이미지가 있고, 컴퓨터가 이해할 수 있는 geometric 연산을 스스로 찾아서 Conv filter 사이즈를 변형해보자

### 학습 방법, Offset
(a) : 우리가 일반적으로 아는 Convolution 연산이다.

근데 각 커널이 바로 뒤에 있는 이미지 픽셀에 곱하는게 아니라,  
각 점마다 다른 픽셀과 곱하라는 Offset을 추가해서 Conv 연산을 해보자.

(b) : 각 픽셀마다(커널의 값마다) Offset을 랜덤으로 부여함.
(c) : 팽창시키는 Offset을 부여해서 이미지를 확대하는 경우에 사용될 수 있음.
(d) : 왼쪽으로 도는 Offset 부여, 왼쪽으로 45도 Rotate하는 geometric transform이다.




offset을 통해서 객체가 있을 법한 위치를 더 잘 학습하게 되고, Conv 연산 결과 유의미한 값이 나오게 된다.
