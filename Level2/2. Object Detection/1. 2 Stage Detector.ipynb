## 1. R-CNN
1. 입력 이미지가 주어지면
2. 객체가 있을 법한 곳에 후보 영역(RoI, region of interest)을 예측
3. 예측된 후보 영역을 고정된 사이즈로 warping한다. (resize)
4. 위의 이미지를 사전 훈련된 CNN 네트워크에 넣어서 vector를 뱉는다.
5. output feature를 SVM에 넣어 분류를 하고, bbox regressor로 좌표값을 예측한다.

### Sliding Window
- 여러가지 window size가 존재한다.
- 윈도우를 이미지 위에 sliding해서 region들을 뽑는다.
- 하지만 이럴 경우 무수히 많은 영역들이 뽑히게 된다.
- 따라서 비추

### Selective Search
- 이미지의 색깔, 색감, shape 등.. 이미지에 존재하는 특성들을 활용함.
- 이미지를 무수히 많은 작은 영역으로 나누고 점점 통합해간다.
- R-CNN에서는 selective search를 통해 RoI를 2000개 추출한다.

## 2. SPPNet
1. 이미지를 Conv layer에 넣는다.
2. 나온 feature map으로부터 다양한 후보영역을 뽑는다. (2000개를 뽑아냄) -> Fast R-CNN에서 자세히 설명함
3. spp 기법으로 고정된 사이즈로 변환한다. (R-CNN의 warping과 다르다.)
4. 이 후의 과정은 R-CNN과 동일


*왼쪽은 R-CNN, 오른쪽은 SPPNet*

### Spatial Pyramid Pooling

- 다양한 사이즈의 RoI가 들어오게 된다.
- 타겟 사이즈를 2x2로 정하고, binning을 한다.
- 각 셀에서 maxpooling을 하든, averagepooling을 하든 하나의 영역에서 feature를 뽑는다.
- 따라서 각 셀마다 하나의 feature를 뽑아내서, 고정된 feature size를 뽑게 된다.

- 최종적으로 concat을 해서 고정된 사이즈의 feature vector를 만들게 된다.
- 최종 vector로 예측을 하게 된다.
- wraping을 하지 않기 때문에 정보의 손실 가능성을 없앨 수 있다.

## 3. Fast R-CNN
1. 이미지를 conv layer에 통과해서 고정된 feature map를 뽑는다.
2. 그로부터 ROI를 뽑는다. (다양한 사이즈가 존재)
3. 여기서 RoI Pooling을 진행해서 고정된 사이즈의 feature vector가 나옴. (spp와 비슷함)
4. classification, regression을 진행함.


### RoI Projection
- feature map상에서 다양한 사이즈의 RoI를 뽑는 방법


1. 원본에서 Selective Search를 통해 2000개를 뽑는다.
2. 원본을 conv에 통과해서 feature map를 만들고,
3. 2000개의 RoI를 feature map에 Projection한다. 
4. 만약 feature map 사이즈가 원본 사이즈보다 작아지면, ROI의 size도 축소해서 projection을 하게 된다.


### RoI Pooling
- 다양한 사이즈의 RoI로부터 고정된 사이즈의 feature로 추출해야 한다.
- SPP와 동일한 방법이지만, 피라미드의 타겟 사이즈를 7x7 하나만 이용한다.

## 4. Faster R-CNN
- Selective Search를 지웠다.
- 그 대신에, RPN이라는 기법을 도입함.
- End-to-End 모델이 되었다.


* 왼쪽은 Fast R-CNN, 오른쪽은 Faster R-CNN

- 이미지를 CNN에 넣어서 feature map를 계산한다.
- 그 feature map을 RPN을 통해 RoI를 뽑아내게 된다.
- 이때, Anchor Box를 사용한다.

### Anchor Box
- convolution feature map으로부터 어떻게 RoI를 뽑아낼까?
- feature map을 그리드로 쪼갠다.
- 각 셀마다 박스 하나씩 포함한다고 본다.
- 근데 이러면 박스 사이즈가 고정이 돼서 원본 사이즈에 mapping 할 때, 다른 사이즈에 대응이 안된다.
- 따라서 각 셀마다 다양한 크기의 사이즈를 정의를 하는데, 이를 Anchor Box라 한다.

- 각 셀마다 N개의 Anchor box를 둔다.
- 객체의 사이즈가 다양해도 어느정도 대응이 가능해진다.

### RPN (Region Proposal Network)
- 하지만 anchor box를 사용하다보니 너무 많은 계산량이 필요해지게 된다.
- 그래서 anchor box가 객체를 포함하는지 확인해야 되고, 만약 포함이 된다면 좌표 또한 미세조정을 해야된다.
- RPN이 이를 도와준다.


- 쉽게 말해, anchor box를 전부 RoI로 사용하면, 개수가 너무 많아지니까 이를 솎아내기 위해 RPN이 있는 것이다.
- 그리고 솎아내기 위해 score가 높은 top N개를 뽑는 역할을 하는 것이다.


- score를 예측하는 classification Head, box prediction Head 총 2가지의 Head가 있다.

### NMS
- RoI를 여러개 뽑다보니, 겹치는 영역이 너무나 많이 생김.
- 서로 겹치는게(IoU) 70% 이상이면 중복된 영역으로 판단해서 제거하게 된다.


1. score로 정렬해서
2. 가장 높은 score를 기준으로 IoU score를 계산한다.
3. 이제 IoU가 0.7 이상이면 지워버린다.
4. 그 다음으로 높은 score를 기준으로 위의 과정을 반복
