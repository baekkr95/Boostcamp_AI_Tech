### 등장 배경
- 보통 모델의 성능을 높이고자 한다면, 모델 사이즈를 키우게 되는데 그렇게 되면 속도가 너무 느려진다.
- 게다가 object detection task의 경우에는 속도가 매우 중요한 평가요소이다.
- 1 stage detector가 빠르기는 하지만 정확도가 낮다는 문제점을 갖고 있다.
- 따라서 자원의 제약이 있는 상태에서 높은 정확도와 효율성을 가진 detection 구조를 만들고 싶음.

## Efficient Det의 challenge
### 1. multi scale feature fusion
- NECK 구조에서 그냥 단순 합을 했다는 점
- 즉, 기존 연구는 resolution 구분 없이 feature map들을 단순 합했다는 것.
- 다른 정보들을 갖고 있는 feature map들을 단순히 합해도 될까?


- 이를 해결하고자 가중합을 사용함.
- BiFPN을 제안했음.
- Cross scale connection 방법을 제안함
    


- 가중합을 할 때, 가중치들을 ReLU을 통과한 값으로 항상 0 이상의 값을 갖는다.

### 2. model scaling
- 기존에는 백본 모델과 이미지 사이즈를 키우기만 했다.
- EfficientNet과 비슷하게 compound scaling을 제안함.
- backbone model로는 EfficientNet B0 ~ B6을 사용함


####  BiFPN 네트워크
- 채널과 깊이를 compound 계수에 따라 증가시킴
- 이미지의 1.35는 grid search로 찾은 값


#### Box/class prediction network
- width는 FPN 맞춰서 고정을 하고,
- depth를 아래 식처럼 증가함.
- 3번째마다 depth가 증가한다.


#### Image resolution
- 선형적으로 증가함
- 기본 사이즈가 512 임
