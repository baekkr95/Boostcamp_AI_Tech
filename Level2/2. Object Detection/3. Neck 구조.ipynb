### Neck 배경



- 기존에는 backbone의 마지막 feature map만 활용을 했었다.
- 중간에 나오는 feature map들도 사용해보자는 의견이 나왔음.

- 물체 인식을 할 때, 크기가 큰 물체와 작은 물체 모두 인식을 해야 된다.
- 근데, Neck을 사용하면 여러 크기의 feature map을 사용하기 때문에 다양하게 인식을 할 수 있게 된다.
- low level의 feature map(초기)은 작은 물체를, high level의 feature map은 큰 물체를 보게 된다.
- low level의 feature는 semantic 정보가 부족해서 high level의 feature map과 섞어주는게 좋다.

## FPN
### Lateral connections

- 상위 level(high level)의 feature를 low level로 전달하기 위해 Lateral connections을 사용한다.
- Top-down pathway의 feature map은 Upsampling을 진행하고
- Bottom-up pathway의 feature map은 1x1 convolution을 진행한다.
- 결과적으로 2개의 feature map shape을 맞춰주고, elementwise하게 더해준다.

## PANet
- FPN의 backbone이 resnet이기 때문에 bottom-up pathway가 엄청 길다.
- low level feature가 high level feature로 제대로 전달이 안 되는 문제점이 있음.


- PANet은 top-down이후에 bottom-up을 하나 더 추가했다.

## DetectoRS
- RFP(Recursive Featrue Pyramid)와 SAC(Switchable Atrous Convolution) 2가지 주요 구성이 있음.
- Neck에서 사용되는 것이 RFP이다.
- 말 그대로 Neck을 Recursive하게 사용한다.


- FPN에서 다시 backbone으로 전달하는 화살표가 있는 것을 볼 수 있다.
- 다시 backbone으로 feature를 전달할 때, ASPP 과정이 포함된다.

### ASPP
- dilated rate를 여러가지로 줘서, receptive field를 점점 키우면서 pooling을 진행한다.
- 이렇게 pooling된 feature들을 concat함.
