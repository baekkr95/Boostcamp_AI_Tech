## Text Detection
- 글자 검출은 이미지에서 읽어낼 글자가 놓여있는 영역을 찾아내는 기술로, OCR 시스템에서 가장 먼저 수행되는 단계


### 일반적인 객체 검출과 다른 점
- 예측하고자 하는 정보가 다르다
- 일반 객체는 클래스와 위치를 예측하지만, 글자 검출은 위치만 검출한다. 왜냐하면 글자라는 단일 클래스만 예측하기 때문임.
- 글자 객체는 일반 객체와 다르게 밀도가 매우 높고, 종횡비가 극단적이다.
- 또한 종이나 옷에 글자가 있는 경우가 많기 때문에, 구겨지거나 휘어지는 등.. 특이 모양이 많이 존재한다.
- 글자는 국가나 광고 종류에 따라 가로쓰기, 세로쓰기가 존재하기도 한다.


### 글자 영역 표현법
1. 직사각형 : x1, y1, w, h
2. 직사각형 + 각도 : x1, y1, w, h, $\theta$
3. 사각형 : x1,y1, x2,y2, x3,y3, x4,y4 (4개의 점을 시계 방향으로 찍음)
4. 다각형 : 왜곡이 있는 글자를 추출하기 위해 사용함. 자연어 후처리를 위해 짝수개의 포인트를 찍는게 좋다.

### Regression based
- 이미지를 입력으로 받아 글자 영역 표현값을 출력으로 한다. (x,y,w,h)
- SSD 모델을 글자 영역 검출 task에 맞게 수정한 **TextBoxes**을 사용함.
- 고정 위치, 고정 박스를 사용한다.
- anchor box별로 실제 글자 영역의 상대 위치와 상대 크기를 추정한다.



- 단점은, 주로 사각형 위주의 글자 표현에만 효과적이다. (Aribitrary-shaped Text 검출이 어려움)
- 또한, anchor box보다 훨씬 큰 종횡비를 갖는 글자는 추출하기 어렵다.

### Segmentation based
- 이미지를 입력으로 받아 글자 영역 표현값들에 사용되는 화소 단위 정보를 뽑고, 후처리를 통해서 최종 글자 영역 표현 값들을 확보한다.
- **PixelLink**을 사용한다.
- 화소 단위로 뽑는 정보가, 각 화소 별로 글자 영역에 속할 확률이다.
- 화소 간의 관계를 뽑는다. 예를 들어 하나의 화소에 대해 주변 8개의 화소도 글자 영역에 속하는지 파악한다.

- 위처럼 화소 단위 정보를 뽑으면, 후처리를 통해 글자 영역 표현 값들을 확보한다.
- Binarization (이진화) : 글자 영역에 속할 확률에 대해 임계치를 적용해서 실제 영역을 뽑는다.
- Connected Component Analysis (연결 성분 분석) : 실제 글자 영역 화소별로 8개 방향으로 이웃한 화소가 글자 영역에 속할 확률값을 사용해서 CCA를 적용하면, 실제 글자들이 뽑아진다.
- RBOX : opencv를 활용해서 RBOX 형태로 글자 형태 표현이 가능해짐


- 단점은, 후처리가 너무 복잡하고 시간이 오래 걸린다.
- 확률값을 사용하는데, 글자 영역이 인접하게 되면 별도의 영역을 하나로 검출할 확률이 높아진다.

## EAST (Efficient and Accurate Scene Text Detector)
- 일종의 segmentation 방법으로 화소 단위 정보를 뽑고 후처리를 통해서 글자 영역을 검출한다.






- 화소 정보로는, 각 화소가 중심에 속하는지에 대한 score map
- 글자 영역 내의 중심에 해당하는지를 확인



- 그리고 화소가 글자 영역이라면 해당 bbox의 위치가 어디인지에 대한 geometry map를 사용



### 모델 구조
- 모델 전체 구조는 U-Net을 사용한다.
- 크게 3가지로 구성된다.
1. Feature extractor stem : 이미지를 입력받아 가장 작은 크기의 feature map까지 추출한다.
2. Feature merging branch : feature map을 키워가며 여러 map을 합쳐간다. 크기를 키우기 위해 unpool 연산을 한다.
3. Output layer : 최종 출력을 뽑는다. 화소 단위 정보를 만들어주는 부분이다.
