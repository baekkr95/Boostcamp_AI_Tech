## OCR (Optical Character Recognition)
- 최근 딥러닝을 적용한 OCR을 STR(Scene Text Recognition)이라 부르기도 한다.
- 글자를 먼저 찾고(영역 감지), 영역 별로 글자를 인지하는, 크게 2가지 조합으로 이해할 수 있다.
- 검출기(영역 감지)는 이미지 전체가 입력이고, 인식기(글자 인지)는 이미지에 있는 글자 영역별로 잘라서 입력으로 사용한다. (둘의 모델 구조를 다르게 설계한다)

## Text Detector
- 글자 영역이냐 아니냐만 다루기 때문에 단일 클래스 문제이다.
- 대신 글자 영역이 여러개 존재할 수 있기 때문에 다수 객체 검출 문제이다.
- 즉, 클래스가 하나인 다수 객체 검출 문제로 볼 수 있다.


- 근데 일반적인 객체 검출과의 차이점이 있다.
- 영역의 종횡비가 일반적인 객체 검출과 많이 다르다. 아무래도 단어, 문장이기 때문에 한글같은 경우에는 가로로 길 수가 있다.
- 그리고 논문 같은 경우에는 객체들의 밀도가 매우 높은 것을 알 수 있다.

### Text Recognizer
- 글자 인식기가 컴퓨터 비전과 자연어 처리의 교집합 영역이다.
- 입력이 이미지고, 출력이 자연어이기 때문이다.
- 그렇기 때문에 보통, 모델의 앞에는 이미지를 다루는 구조가 들어가고 모델의 뒷부분은 자연어를 처리하는 모델 구조가 들어가게 된다.
- 비슷한 기술로 Image Captioning이 존재한다.



### Serializer
- 글자 인식기(Recognizer)의 결과값을 일렬로 쭉 정렬시키는 과정이 있어야 한다.
- 이를 통해 자연어 처리 모듈을 붙일 수가 있다. (금칙어 처리, 요약, 의미 파악 등..)
- 사람이 읽는 순서대로 정렬하는 것이 보편적이다.




### Text Parser
- Serializer에서 말한 자연어처리 모듈이다.
- 기 정의된 key들에 대한 value를 추출하게 된다.

### BIO Tagging
- Parser에서 Key와 Value를 추출할 때 주요한 접근론 중 BIO 태깅이 있다.
- 이미 정의된 개체(사람명, 주소, 이름 등..)에 대한 값을 추출한다.
