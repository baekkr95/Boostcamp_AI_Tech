### 내 ip를 알아내는 방법
1. `ip addr` 명령어
2. `curl ipinfo.io/ip` : ipinfo.io/ip 사이트에서 알아내는 방법

### 근데 2개의 ip 결과가 다를까? (같은 경우도 있다)
- 인터넷 계약을 맺고 여러 디바이스에 인터넷 연결하기 위해 Router를 사용한다.
- 외부에 공개되는 ip를 public address
- 그리고 이 public address는 Router의 ip가 된다.
- 외부에 공개되지 않고, 자체적으로 사용되지 않는 ip를 private address라고 한다.
- 그래서 Router 없이 통신사와 계약한 회선이 바로 디바이스에 꽂히면 위의 ip 알아내는 방법의 2가지 ip 결과가 같게 된다.

## SSH
- 서버를 쉘을 통해 제어를 하기 위해, SSH를 사용한다.
- 제어하고 싶은 대상(서버)에 SSH Server를 깔고, 
- 서버를 제어하려고 하는 내 컴퓨터(클라이언트)에는 SSH Client를 깔면 된다.
- 그래서 SSH Client가 명령어를 서버에 전달을 하고, 서버는 그 결과를 Client에 출력한다.
- 그리고 유닉스 계열 시스템에서는 SSH가 기본적으로 설치되어 있다. 따로 깔 필요가 없음


- `ssh 접속서버 아이디@접속서버 아이피` 를 통해 원격 접속

## Port

- 웹 브라우저는 80포트에 기본적으로 연결하도록 한다.
- SSH는 서버는 22번 포트를 기본으로 연결하도록 한다.
- 포트는 0번부터 65000개의 포트가 있다.
- 즉, 웹브라우저에서 웹서버 주소를 입력할 때 특별히 포트번호를 적지 않는다면, 서버 컴퓨터의 80번 포트를 찾는 것임
- 0번에서 1024번까지는 well-known port(표준), 나머지 port는 따로 프로그램에 할당을 해준다.


- `sudo nano /etc/ssh/sshd_config`명령을 통해 파일을 열어보면, port 22라고 적혀있는 것을 볼 수 있다.

## Port Forwarding
- 외부 사용자가 Private ip에 접근할 수는 없을까?
- 이때, port forwarding을 사용한다
- 라우터에 존재하는 포트를 사용해서, 사용자가 특정 포트를 사용해서 접근이 가능함

### default gateway
- 포트 포워딩을 위해서, 공유기의 환경을 바꾸는데, 
- 공유기 웹서버에 접근해서 설정해야 함.
- 이때, 내 로컬 컴퓨터가 공유기에 접속하기 위해, 내부에서만 통용되는 ip를 사용한다.
- 이를 default gateway라 한다.

- `ip route`를 사용하면 default gateway의 ip를 알 수 있다.
- 해당 ip를 입력하면 라우터의 관리자 화면으로 이동한다.
- `관리도구` -> `고급설정` -> `NAT` -> `포트포워드 설정`로 이동하면, 
- `내부 IP주소`가 있고, 이는 내가 서버를 설치하고 싶은 컴퓨터의 내부 ip이다.
- `외부 포트`는 외부 사용자가 접근할 때 사용하는(라우터로 들어올때) 포트 번호이다.
- `내부 포트`는 컴퓨터 상의 포트이다. 컴퓨터에 설치된 웹서버가 몇 번째 포트에서 listening을 하는가


- 이제 내 서버에 접속할 사용자들에게, 라우터의 ip와 포트 번호를 알려주면 된다.
- 내 공개 ip는 위에서 사용한 `curl http://ipinfo.io/ip`로 알 수 있다.
- 그리고 사용자가 공용 ip에 포트 번호 9000을 입력하면 된다.
