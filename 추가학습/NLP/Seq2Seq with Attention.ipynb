## Seq2Seq Model

- many to many 문제에 해당된다.
- 입력 문장이 들어오면, 출력 문장이 들어온다.
- 인코더와 디코더로 이루어진다.
- 인코더와 디코더는 서로 파라미터를 share하지 않는다.
- RNN 모델로써 LSTM을 사용했다.
- 인코더 마지막 time step의 hidden state vector는 디코더의 첫번째 time step의 입력이 된다. (디코더의 $h_0$)


- 첫번째 단어에 특수문자로써 `<sos>`를 사용한다.
- 문장이 끝나는 시점은 마지막에 `<eos>`를 사용한다. (eos 토큰이 나오면 생성을 종료함)


## Seq2Seq Model with Attention
- RNN의 특성상, 앞에 나왔던 단어들을 고정된 벡터안에 전부 압축해서 저장을 하게 된다.
- 그리고 이전의 정보들을 점점 소실하거나 변질될 수 있다.


- Attention module을 사용하자.
- 디코더가, 인코더의 마지막 hidden state vector에만 의존하지 않아도 된다.
- 입력 문장에서 나오는 hidden state vector들을($h_0, h_1, ...)$ 디코더가 단어를 생성할 때 필요로 하는 인코더의 hidden state vector를 적절히 선택해서 예측에 활용한다.

---
### 예시
- 프랑스어를 영어로 번역
- 인코더의 hidden state vector가 4개 있다.


- 마지막 time step의 h_4가 디코더의 h_0로써 입력으로 들어간다.
- 디코더는 sos 토큰의 word embedding vector가 입력으로 주어진다.
- 위의 2가지를 입력으로 받아서 디코더의 hidden state vector h_1이 생긴다.


- 이제 디코더의 h_1을 가지고 다음 단어의 예측에 사용을 하고,
- 이것을 통해 인코더의 4가지 hidden state vector 중 어떤 것을 필요로 할지 선정을 한다.


- 인코더 hideen state 각각과 내적을 한다.
- 이 내적 값을 유사도라고 볼 수 있다.
- 그리고 이 내적 값을 softmax를 통과시키면 확률 값이 나온다. (attention vector)
- 이 확률 값은 인코더의 hideen state vector에 부여되는 가중치로써 사용이 된다.


- 따라서, 인코더의 hideen state vector와 확률값 간에 가중평균을 만들 수 있고, 이 가중평균으로써 나오는 하나의 인코딩 벡터를 구하게 된다.
- 가중평균으로써 나온 이 벡터는 context vector라고도 부른다. (attention module의 output)


- 여기까지가 지금 디코더의 hideen state vector h_1을 사용한 것임.
- 그림에서는 디코더의 h_1이 인코더의 h_1를 85%, h_2를 8%...의 정보를 필요로 하고 있다는 것이다.

- 디코더 hideen state vector와 context vector(attention output)가 같이 concat이 되고,
- output layer의 입력으로 들어가고, 다음 나올 단어를 예측하게 된다.
