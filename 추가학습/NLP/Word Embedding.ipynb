## Word Embedding
- 단어들을 특정한 차원으로 이루어진 공간상의 점 또는 벡터로 전환하는 것
- 좌표 공간의 차원 수를 사전에 미리 정의해서 워드 임베딩
- 해당 좌표 공간 상에서 각 단어에 대한 최적의 좌표 값 또는 벡터 표현을 출력으로 함.
- 비슷한 단어를 비슷한 위치의 점으로 mapping 되도록 한다.
- 따라서 단어들의 의미상의 유사도를 반영할 수 있다.
- 워드 임베딩을 통해서 의미를 잘 반영된 벡터를 학습하면 이를 입력으로 받아서 NLP를 수행할 수 있다.

## Word2Vec
- 워드 임베딩하는 학습 방법 중 가장 유명한 방법
- 같은 문장에서 인접한 단어들 간에 의미가 비슷할 것이라는 가정을 사용함.
- 한 단어가 주변의 단어를 통해 의미를 알 수 있다.
- 원하는 단어 주변에 나타나는 단어들의 확률 분포를 예측하게 된다.


### 예시
- ex) I study math
- ex) 임베딩 차원을 2로 세팅
- 워드별로 구분하는 tokenize 과정을 수행해야 된다.
- unique한 단어만을 모아서 사전을 구축한다. (vocabulary)
- 사전의 각 단어는 사전의 사이즈만큼의 차원을 가지는 one hot vector를 갖는다. (예시에서는 3차원)
- sliding window를 사용해서 단어 앞뒤를 살펴본다.
