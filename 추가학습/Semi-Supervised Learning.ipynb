## Semi-Supervised Learning
- 준지도 학습
- 목적 : Unlabeled Data를 어떻게 활용하면, supervised learning의 퍼포먼스를 향상시킬 수 있을까?


- 그림 C의 분류 경계선과 D의 분류 경계선의 기울기가 다르다.
- Unlabeled data들이 기울기를 바꾸게 함.
- 정답 정보는 없더라도, 데이터의 분포를 좀 더 자세하게 고려할 수 있게 한다.
- 그래서 소수의 labeled된 소수의 data만 가지고 supervised learning만 하는 것보다 성능이 좋지 않을까 기대할 수 있는 것.

### 전제조건
- 어떤 상황이어야 Unlabled data가 유효할까?
- Y라는 class 정보들에 따라서 데이터가 어느정도 분포를 갖고 있어야함.
- 데이터가 완전히 랜덤 데이터면 의미가 없어짐

### Semi-Supervised vs Transductive learning
- Semi-Supervised : 준지도 학습은 지금 현재 알 수 없어도, 새로운 test data에 대한 x정보가 주어졌을 때, 그 시점에서의 Y를 잘 예측하자


- Transductive learning : 지금 현재 주어진 unlabeled data에 대한 Y를 잘 예측하자

### Self-Training
- labeled data에 대해 학습을 하고 (Function $F$)
- unlabeled data에 적용을 해서, 예측을 함
- 그것을 pseudo-labeled data라 칭함.
- 마지막에 labeled data와 pseudo-labeled data를 합쳐서, retraining 한다. (Function $G$)


- 장점은 단순하고 구현이 빠른 기법이다.
- 단점은 처음 pseudo-label에서 실수가 나오면 모델 자체가 망함

### Co-Training, 협업 학습
- 어떤 특정 개체(class)를 표현하는데 있어서, 상호 배반적인 feature가 있을 것이라 가정함.
- class를 표현하는데 있어서, 이미지라는 feature set과 텍스트라는 feature set은 굉장히 다를 것이다. feature set이 mutually exclusive하다는 가정을 한다.
- 완벽히 mutually exclusive 하지 않을 가능성이 높다. 그럼에도 그 가정을 조금 위배해도 서로 다른 feature set이 만들어진다면 그것을 사용하자는 것.


- multi view 알고리즘 이라고도 함.
- x를 image feautre, text feature 두 가지로 split한다. (feature split)
- 각 feature를 classification하는 모델을 만든다.
- 그 다음에, 한 쪽에서 자신있게 분류한 분류기를 다른 쪽의 분류기를 학습하는데 사용한다.

- 1번 분류기에 의해 만들어진 pseudo-labeled dataset이 2번 모델을 만드는 학습 데이터에 추가가 된다.
