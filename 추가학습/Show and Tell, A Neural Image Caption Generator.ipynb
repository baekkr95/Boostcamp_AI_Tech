### Image Caption Generator (이미지 캡션 생성)
- 이미지를 설명하는 문장을 생성하는 기술
- 대표적인 모델로, Neural Image Caption (NIC)가 있다.
- 논문 링크 : https://arxiv.org/abs/1411.4555

## Neural Image Caption, NIC
- 기계 번역 작업에서 인코더(encoder) 부분을 CNN으로 대체했다.
- CNN 네트워크를 이용해서 이미지의 특징을 추출하고 RNN(LSTM 사용)을 거쳐 문장을 생성함.


- CNN으로 feature tensor를 뽑아서 이를 fixed size인 context vector를 만들고, 이를 Decoder 넣는다.
- Decoder(LSTM)는 이미지를 설명하는 내용을 단어마다 뽑아내서 결과 문장을 만든다.
- 물론 Transformer 이후로는 이렇게 context vector를 뽑는 방법은 거의 사용하지 않는다.


- 입력 : Image ($I$)
- 출력 : Target Sentence ($S = \{S_1, S_2,...,S_n\}$)
- 가능도 $P(S|I)$를 최대화하는 문제로 정의가 가능 (어떠한 이미지가 주어졌을 때, 어떤 문장이 더 그럴싸한지 예측할 수 있도록 함)
- 실제 학습할 때, 각 이미지마다 그에 따른 정답 문장을 pair로 마련하고, 어떠한 입력 이미지가 들어왔을 때 정답 문장을 출력하는 형태로 모델을 학습한다. 즉, 이미지가 주어졌을 때 정답 문장의 확률값을 maximize하는 형태로 학습함.

### 공식
- 하나의 이미지를 정답 문장으로 번역하는 작업으로 보면 된다.


- $\theta$ : Conv filter들의 parameters, Lstm의 가중치 값
- $log p(S|I)$ : 어떤 입력 $I$가 주어졌을 때, 거기서 예측된 $S$값의 확률값의 Log Likelihood
- 로그 가능도 값을 maximize하는 파라미터를 찾는다.
- 학습할 때, 이미지와 문장을 pair로 묶기 때문에 $\Sigma$ 밑에 $(I,S)$가 있는 것을 알 수 있다

- CNN 모델에서 나온 feature vector를 첫번째 LSTM에 hidden state의 초기값으로 설정
- 첫번째 input word $S_0$를 generation하고, 다음 word의 확률을 계산
- 얻은 확률의 log값을 계산하고, 이렇게 나온 모든 log값을 모두 더하면 loss function이 된다. (부호 바꿔서 minimize)

### Beam Search
- 항상 가장 확률이 높은 K개의 candidates를 갖도록 한다. (여기선 20개 사용)
- 마지막에는 한개로 추린다.
- 단순히 가장 확률이 높은 단어 하나를 선택하지 않고, 상위 K개를 뽑아서 그 단어들로 다음 단어를 예측하는 듯? 그래서 각 단어들의 확률을 더해서 문장 단위로 가장 높은 확률의 문장을 결과로 쓴다,
