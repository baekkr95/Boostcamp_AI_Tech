### 생성 모델 (Generative Model)
- 생성 모델은 있을 법한 이미지를 생성하는 모델이다.
- 원래 이미지 데이터의 분포를 근사하는 모델을 만드는 것이 목표이다.
- 학습을 하면서 생성 모델 G가 원본 데이터의 분포를 학습하는 것
- 즉, 학습이 잘 되었다는 것은 원본 데이터의 분포에 근사할 수 있다는 것이고 통계적으로 평균적인 특징을 갖는 데이터를 생성할 수 있다.
- 최종적으로 판별자가 이미지의 진짜, 가짜 판별이 안되기 때문에 1/2이 나온다

## GAN (Generative Adversarial Networks)
- 생성자 G, 판별자 D 두개의 네트워크를 활용한 생성 모델
- 학습 완료 후에 사용하는 것이 생성자 G 모델이다. 판별자는 D는 생성자가 학습을 잘 하도록 도와줌

#### `목적함수`
- 목적 함수 V는 D와 G로 구성된다.
- G는 V 값을 낮추고자, G는 V를 높이고자 한다. (min max 최적화)

#### `판별자 D`
- $p_{data}$는 원본 데이터의 분포를 의미한다.
- D는 이미지 x를 받아서 그 이미지가 얼마나 진짜 같은지에 대한 확률값을 출력함. (0 ~ 1)
- $E_{x\sim~p_{data(x)}} {[logD(x)]}$ : 여러개의 데이터(x)를 뽑고 log D에 넣고 평균값을 구한다

#### `생성자 G`
- 생성자는 항상 노이즈 vector인 z(latent vector)로 부터 입력을 받아서 새로운 이미지를 만든다.
- $E_{z\sim~p_{z(z)}} {[log(1-D(G(z)))]}$ : 하나의 노이즈를 샘플링해서 이를 생성자 G에 넣어 가짜이미지를 만든다. 그 가짜 이미지를 D에 넣고 log(1-D) 값의 평균값을 구한다.

#### `결론`
- 판별자는 원본 데이터에 대해서 1로 분류할 수 있도록 학습한다.
- $logD(x)$를 maximize한다는 것은 원본 데이터에 대해서는 1을 출력할 수 있도록 학습하겠다는 것이다.
- 반면에 가짜 이미지가 들어왔을 때는($log(1 - D(G(x))$), 0을 출력할 수 있도록 학습하겠다는 것이다.


- 생성자 G는 오른쪽 수식 $E_{z\sim~p_{z(z)}} {[log(1-D(G(z)))]}$을 minimize 하기 때문에, 자신이 만든 가짜 이미지를 D에 넣었을 때 1을 출력할 수 있도록 하는 것이 목표다. (그럴싸한 이미지로 보이도록)
