## Self-Supervised Learning & Contrastive Learning

- Self-Supervised Learning : 사람이 label를 하나하나 구현하기 어려워서, 데이터 자체에서 쉽게 얻을 수 있는 정보를 Label로 사용하자


### Contrastive learning 배경
- Leopard에 대한 예측 확률 값을 보니, 재규어와 치타와 비슷한 클래스의 확률 값이 다른 class들에 비해 훨씬 높은 것을 볼 수 있다.
- laebl 정보와 상관없이, 이미지끼리 공유하는 어떤 특징이 있을거라고 생각이 들었음. (instances 간 유사도가 있다?)
- 관측치 사이의 유사성을 기반으로, 관측치끼리 구분하도록 학습하면 label 정보가 없어도 이미지를 분류하는 representation을 얻을 수 있을 것 같다. 

### Positive, Negative Pair
- Positive pair 끼리는 거리를 좁히고, Negative pair끼리는 거리를 멀리 띄워놓는 것이 학습 원리임


- 강아와 의자 각각 augmentation하고, augmentation된 이미지들끼리는 positive pair로써 거리를 좁히도록 학습
- 하지만 강아지와 의자끼리는 negative pair로써 거리를 벌리도록 학습함
